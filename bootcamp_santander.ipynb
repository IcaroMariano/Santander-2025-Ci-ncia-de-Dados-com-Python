```python
#EXTRACT
import pandas as pd
df = pd.read_csv("santander_bootcamp.csv")
df
```

```python
#TRANSFORM
def gerar_mensagem(nome, saldo):
    if saldo < 500:
        return f"{nome}, que tal começar a investir aos poucos para fazer seu dinheiro crescer?"
    elif saldo < 1500:
        return f"{nome}, você já tem uma boa base financeira. Investir pode te levar mais longe."
    else:
        return f"{nome}, excelente controle financeiro! Invista para potencializar seus resultados."
df["news"] = df.apply(
    lambda row: gerar_mensagem(row["Name"], row["Balance"]),
    axis=1
)
```

```python
#LOAD
df.to_csv("santander_resultado.csv", index=False)
from google.colab import files
files.download("santander_resultado.csv")
```
